```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

#### OpidsVA

```python
df_op = pd.read_csv(r'C:\Users\babbra\CTBA\M2\OpidsVA.csv')
df_op
df_op.head()

plt.scatter(df_op['FPOO-Rate'],df_op['MedianHouseholdIncome'])
plt.title('OpioidsVA Scatterplot')
plt.xlabel('Opioid Overdoses Per 100,000 County Residents')
plt.ylabel('Median Household Income')
plt.show()
### This plot accurately displays how opioid overdoses in VA are correlated to individuals with low income.

plt.scatter(df_op['FPOO-Rate'],df_op['FFHO-Rate'])
plt.title('Opioids Versus Heroin Overdoses')
plt.xlabel('Opioid Overdoses Per 100,000 County Residents')
plt.ylabel('Heroin Overdoses Per 100,000 County Residents')
plt.show()
### This plot shows how there are many counties with a lower number of oversdoses, but there are a select few that have more than 25 per 100,000 county residents.
```

#### Presidents

```python
df_pres = pd.read_csv(r'C:\Users\babbra\CTBA\M2\presidents.csv')
df_pres
df_pres.head()

plt.scatter(df_pres['Location'],df_pres['State'])
plt.title('US Presidents Birthplaces')
plt.xlabel('Birthplace Location')
plt.ylabel('State')
plt.show()
### This plot shows that Virginia and Ohio have the greatest number of US presidents in terms of their birthplaces.
```
